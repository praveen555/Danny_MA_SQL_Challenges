## 9.What was the total volume of pizzas ordered for each hour of the day?

it makes more sense to perform analytics on those orders which were not cancelled.

```python
df_merge=pd.merge(df_cust_orders,df_runner_orders,on="order_id")
df_merge=df_merge[df_merge['cancellation'].isna()]
df_merge
```

```python
df_merge['order_time']=pd.to_datetime(df_merge['order_time'])
```

```python
df_merge['date']=df_merge['order_time'].dt.date
df_merge['time']=df_merge['order_time'].dt.time
df_merge['hour']=df_merge['order_time'].dt.hour
```

```python
df_x=df_merge.groupby(['date','hour'])['hour'].count().to_frame("count_hour").reset_index()
```

```python
df_x['vol']=df_x['count_hour']/sum(df_x['count_hour'])*100
df_x
```

##10.What was the volume of orders for each day of the week?

```python
df_merge=pd.merge(df_cust_orders,df_runner_orders,on="order_id")
#df_merge=df_merge[df_merge['cancellation'].isna()]
df_merge
```

```python
df_merge['order_time']=pd.to_datetime(df_merge['order_time'])
df_merge['date']=df_merge['order_time'].dt.date
df_merge['time']=df_merge['order_time'].dt.time
df_merge['day_of_week']=df_merge['order_time'].dt.dayofweek
```

```python
df_merge
```

```python
df_x=df_merge.groupby(['day_of_week'])['order_id'].count().to_frame("count")
```

```python
df_x['vol']=df_x['count']/sum(df_x['count'])*100
df_x
```
